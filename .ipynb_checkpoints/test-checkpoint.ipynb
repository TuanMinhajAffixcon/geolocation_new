```python
import numpy as np
import streamlit as st
import pandas as pd
import geohash2

def generate_random_coordinates(lat, lon, distance, num_points):
    """
    Generate random latitude and longitude coordinates within a specified distance from a given point.
    
    Args:
    - lat (float): Latitude of the center point.
    - lon (float): Longitude of the center point.
    - distance (float): Maximum distance from the center point (in kilometers).
    - num_points (int): Number of random coordinates to generate.
    
    Returns:
    - List of tuples: Each tuple contains a pair of latitude and longitude coordinates.
    """
    coordinates = []
    while len(coordinates) < num_points:
        # Generate random latitude and longitude offsets
        lat_offset = np.random.uniform(-1, 1)
        lon_offset = np.random.uniform(-1, 1)
        
        # Calculate new latitude and longitude
        lat_new = lat + lat_offset * (distance / 111)
        lon_new = lon + lon_offset * (distance / (111 * np.cos(np.radians(lat))))
        
        # Calculate distance from center point
        dist = np.sqrt(lat_offset**2 + lon_offset**2) * distance
        
        # Add coordinates if within the specified distance
        if dist <= distance:
            coordinates.append((lat_new, lon_new))
    
    return coordinates

# Given center point
center_lat = -33.893000
center_lon = 151.215000

# Maximum distance from the center point (in kilometers)
max_distance = 0.022

# Number of random coordinates to generate
num_points = 80

# Generate random coordinates within 100 km
random_coordinates = generate_random_coordinates(center_lat, center_lon, max_distance, num_points)

# Convert coordinates to DataFrame
df = pd.DataFrame(random_coordinates, columns=['Latitude', 'Longitude'])

# Convert coordinates to geohashes for workgeohash and homegeohash
df['workgeohash'] = df.apply(lambda row: geohash2.encode(row['Latitude'] + 0.1, row['Longitude']), axis=1) # Example: Adding 0.1 to latitude for workgeohash
df['homegeohash9'] = df.apply(lambda row: geohash2.encode(row['Latitude'] - 0.1, row['Longitude']), axis=1) # Example: Subtracting 0.1 from latitude for homegeohash

# Save DataFrame as CSV
df.to_csv('loc5_workgeo_homegeo.csv', index=False)
# df
```

```python
from collections import Counter

items = {'loc1': ['A', 'B', 'C', 'D','P'],
         'loc2': ['A', 'B', 'E'],
         'loc3': ['B', 'C', 'X'],
         'loc4': ['C'],
         'loc5': ['A', 'C', 'Z']}

# Create a counter for all items
all_items_counter = Counter([item for sublist in items.values() for item in sublist])

# Find records that are unique to each location
unique_records_per_location = {}
for location, records in items.items():
    location_counter = Counter(records)
    unique_records_per_location[location] = [item for item in records if location_counter[item] == 1 and all_items_counter[item] == 1]

locations_list = []
common_items_list = []
item_count_list = []
# Print unique records for each location
for location, unique_records in unique_records_per_location.items():
    locations_list.append(location)
    common_items_list.append(unique_records)
    item_count_list.append(len(unique_records))
# Create the DataFrame
df_common_items_counts = pd.DataFrame({
    'Locations': locations_list,
    'Unique Items': common_items_list,
    'Count': item_count_list
})

df_common_items_counts=df_common_items_counts[df_common_items_counts['Count']>0]

# Display the DataFrame
(df_common_items_counts)
```

```python
import pandas as pd

# Initialize lists to store data
locations_list = []
common_items_list = []
item_count_list = []

# Iterate through the common_items_counts dictionary
for combination, counts in common_items_counts.items():
    # Extract the locations, common items, and their counts
    locations = ', '.join(combination)
    common_items = ', '.join([item for item, count in counts.items() if count > 0])
    item_count = sum(counts[item] for item in counts if counts[item] > 0)
    
    # Append the data to the lists
    locations_list.append(locations)
    common_items_list.append(common_items)
    item_count_list.append(item_count)

# Create the DataFrame
df_common_items_counts = pd.DataFrame({
    'Locations': locations_list,
    'Common Items': common_items_list,
    'Count': item_count_list
})

df_common_items_counts=df_common_items_counts[df_common_items_counts['Count']>0]

# Display the DataFrame
(df_common_items_counts)
```
